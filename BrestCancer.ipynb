```python
'''importing pandas'''
import pandas as pd
```

```python
'''Reading csv file using pandas'''
df = pd.read_csv('BrestCancer.csv')
```

```python
df.head()
```

```python
'''Droping unnecessary columns'''
df.drop(['id','Unnamed: 32'], axis= 1, inplace= True)
df.head()
```

```python
df.head()
```

```python
'''Spliting data into Trget and dependent column'''
X=df.drop('diagnosis',axis=1,inplace=False)
y =df['diagnosis']
y.head()
```

```python
X.head()
```

```python
'''Importing preprocessing to encode categorical feature to numerical feature'''
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
le.fit(y)
y=le.transform(y)
```

```python
y
```

```python
'''Splittinf dataset into train and test dataset'''
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=9)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train
```

```python
'''Importing Random Forest Classifier to classify dataset'''
from sklearn.ensemble import RandomForestClassifier
```

```python
'''Fitting dataset to Random Forest '''
classifier = RandomForestClassifier(n_estimators=10).fit(X_train, y_train)
```

```python
cls_score = classifier.score(X_test, y_test)
cls_score
```

```python
'''Testing model on test dataset'''
from sklearn.metrics import confusion_matrix,recall_score,f1_score,precision_score
```

```python
y_test_pred = classifier.predict(X_test)
```

```python
confusion_matrix(y_test, y_test_pred)
```

```python
recall_score(y_test,y_test_pred)
```

```python
f1_score(y_test,y_test_pred)
```

```python
precision_score(y_test,y_test_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
 print(classification_report(y_test, y_test_pred,target_names=['B','M']))
```
